```python
#crime data record of 2001-2012:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data1 = pd.read_csv("crime_(2001-2012).csv")
data1 = data1.rename(columns = {'CRIME HEAD': 'CRIME_HEAD'})

data1  = data1.drop(data1[(data1['STATE/UT'] == 'ALL INDIA ') | 
                    (data1['STATE/UT'] == 'ALL INDIA K&A') |
                    (data1['STATE/UT'] == 'All India') |
                    (data1['STATE/UT'] == 'TOTAL Crime against Women')].index)


#np.unique(data1['CRIME_HEAD'].tolist())

data1
```

```python
#crime data record of 2013:
data2 = pd.read_csv("crime_2013.csv")
data2 = data2.rename(columns = {'Crime Head': 'CRIME_HEAD',
                                'State/UTs' : 'STATE/UT'})

data2 = data2.drop(data2[(data2['CRIME_HEAD'] == 'Commission of Sati (P) Act') |
                         (data2['CRIME_HEAD'] == 'Importation of Girls from Foreign Country') |
                         (data2['CRIME_HEAD'] == 'Dowry Prohibition Act') |
                         (data2['STATE/UT'] == 'Total (All-India)') |
                         (data2['STATE/UT'] == 'Total (State)') |
                         (data2['STATE/UT'] == 'Total (UTs)')|
                         (data1['CRIME_HEAD'] == 'IMMORAL TRAFFIC (P) ACT')|
                         (data1['CRIME_HEAD'] == 'INDECENT REPRESENTATION OF WOMEN(P) ACT')].index)

data2['CRIME_HEAD'] = data2['CRIME_HEAD'].str.upper()

data2['CRIME_HEAD']=data2['CRIME_HEAD'].str.replace('ASSAULT ON WOMEN WITH INTENT TO OUTRAGE HER MODEST','ASSAULT ON WOMEN WITH INTENT TO OUTRAGE HER MODESTY')
data2['CRIME_HEAD']=data2['CRIME_HEAD'].str.replace('CRUELTY BY HUSBAND OR HIS RELATIVES','CRUELTY BY HUSBAND OR RELATIVES')
data2['CRIME_HEAD']=data2['CRIME_HEAD'].str.replace('DOWRY DEATHS','DOWRY DEATH')                                                  
data2['CRIME_HEAD']=data2['CRIME_HEAD'].str.replace('KIDNAPPING & ABDUCTION OF WOMEN & GRILS','KIDNAPPING & ABDUCTION')  
                                                  
                                                

#np.unique(data2['CRIME_HEAD'].tolist())
#data2 = data2.sort_values('State/UTs')

data2
```

```python
#merging 2 datasets and performed some visualization
crime_df = pd.merge(data1,data2, how = 'inner')

#crime_df[crime_df['CRIME_HEAD']== 'TOTAL CRIMES AGAINST WOMEN']
crime_df = crime_df.iloc[:210,:]
#np.unique(crime_df['STATE/UT'].tolist())


#crime record of ASSAULT ON WOMEN WITH INTENT TO OUTRAGE HER MODESTY
crime_df1 = crime_df[crime_df['CRIME_HEAD']=='ASSAULT ON WOMEN WITH INTENT TO OUTRAGE HER MODESTY'].set_index('STATE/UT')
del crime_df1['CRIME_HEAD']
crime_df1 = crime_df1.T.reset_index()

#crime record of CRUELTY BY HUSBAND OR RELATIVES
crime_df2 = crime_df[crime_df['CRIME_HEAD']=='CRUELTY BY HUSBAND OR RELATIVES'].set_index('STATE/UT')
del crime_df2['CRIME_HEAD']
crime_df2 = crime_df2.T.reset_index()

#crime record of DOWRY DEATH
crime_df3 = crime_df[crime_df['CRIME_HEAD']=='DOWRY DEATH'].set_index('STATE/UT')
del crime_df3['CRIME_HEAD']
crime_df3 = crime_df3.T.reset_index()

#crime record of INSULT TO THE MODESTY OF WOMEN
crime_df4 = crime_df[crime_df['CRIME_HEAD']=='INSULT TO THE MODESTY OF WOMEN'].set_index('STATE/UT')
del crime_df4['CRIME_HEAD']
crime_df4 = crime_df4.T.reset_index()

#crime record of KIDNAPPING & ABDUCTION
crime_df5 = crime_df[crime_df['CRIME_HEAD']=='KIDNAPPING & ABDUCTION'].set_index('STATE/UT')
del crime_df5['CRIME_HEAD']
crime_df5 = crime_df5.T.reset_index()

#crime record of RAPE
crime_df6 = crime_df[crime_df['CRIME_HEAD']=='RAPE'].set_index('STATE/UT')
del crime_df6['CRIME_HEAD']
crime_df6 = crime_df6.T.reset_index()


crime_df6
```
